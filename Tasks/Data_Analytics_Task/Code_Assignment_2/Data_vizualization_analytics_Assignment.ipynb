```python
# Importing required libraries.
import pandas as pd
import numpy as np
import seaborn as sns #visualisation
import matplotlib.pyplot as plt #visualisation

df = pd.read_csv("data.csv")
df
```

```python
df['make']
```

```python
#Question 1
#dropping duplicates from the loaded dataframe
df1=df.drop_duplicates()
df1
```

```python
#Question 2
#dropping rows whose columns have null value

df2=df1.dropna()
df2
```

```python
#Question 3
#method-1 for removing outliers 
column_names=list(df2)
column_names.remove('Make')
column_names.remove('Model')
column_names.remove('Transmission Type')
column_names.remove('Driven_Wheels')
column_names
```

```python
#Function for removal of outliners 
def remove_outliers(df,columns,n_std):
    for col in columns:
        print('Working on column: {}'.format(col))
        
        mu = df[col].mean()
        sd = df[col].std()
        
        df = df[(df[col] <= mu+(n_std*sd))]
        
    return df
```

```python
df3=remove_outliers(df2,column_names,2)   #df3 is data Frame without duplicates, null_values, outliers.       
df3
```

```python
#Question-4 Data Visualization
#boxplot showing most of the outliers removed
sns.boxplot(x=df3['MSRP'])
plt.show()
```

```python
#Histogram for different car manufacturing companies producing cars 
df3.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(12,8))
plt.title("Number of cars by company")
plt.ylabel('Number of cars')
plt.xlabel('Company')
plt.grid()
plt.show()
```

```python
#Histogram for different car manufacturing companies producing cars 
df3.Make.value_counts().nlargest(100).plot(kind='bar', figsize=(12,8))
plt.title("Number of cars by company")
plt.ylabel('Number of cars')
plt.xlabel('Company')
plt.grid()
plt.show()
```

```python
#scatter plot to show corelation
fig, ax = plt.subplots(figsize=(20,10))
ax.scatter(df3['Engine HP'], df3['MSRP'])
ax.set_xlabel('Engine HP')
ax.set_ylabel('MSRP')
plt.show()
```
